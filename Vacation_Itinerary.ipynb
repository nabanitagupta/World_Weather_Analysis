```python
weather_vacy_df1 = weather_vacy_df.loc[(weather_vacy_df["City"] == 'Galveston')]
weather_vacy_df2 = weather_vacy_df.loc[(weather_vacy_df["City"] == 'Hays')]
weather_vacy_df3 = weather_vacy_df.loc[(weather_vacy_df["City"] == 'Laredo')]
weather_vacy_df4 = weather_vacy_df.loc[(weather_vacy_df["City"] == 'China')]
```

```python
import gmaps
import gmaps.datasets
# Import the API key.
from config import g_key
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
# Add a map of the vacation spots and marker layer for each city.
# Latitude-longitude pairs
Galveston = (29.37, -94.97)  
Hays = (30.05, -98.03) 
Laredo = (27.51, -99.51)
China = (25.7, -99.23)
fig = gmaps.figure() 
galveston2china = gmaps.directions_layer(Galveston, China, waypoints=[Hays, Laredo], travel_mode = 'DRIVING')
fig.add_layer(galveston2china)
# Call the figure to plot the data.
fig
```

```python
preferred_city = weather_vacy_df1.copy()
preferred_city = preferred_city.append(weather_vacy_df2)
preferred_city = preferred_city.append(weather_vacy_df3)
preferred_city = preferred_city.append(weather_vacy_df4)
preferred_city = preferred_city.reset_index()
preferred_city = preferred_city.drop('index', 1)
preferred_city = preferred_city.drop('City_ID', 1)
preferred_city
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in preferred_city.iterrows()]
```

```python
# Add a map of the vacation spots and marker layer for each city and a pop-up marker for each city.
locations = preferred_city[["Lat", "Lng"]]
fig = gmaps.figure()
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)
# Call the figure to plot the data.
fig
```
